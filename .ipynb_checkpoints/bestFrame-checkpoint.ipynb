```python
def setDisplay():
    # set display defaults
    plt.rcParams['figure.figsize'] = (10, 10)        # large images
    plt.rcParams['image.interpolation'] = 'nearest'  # don't interpolate: show square pixels
    plt.rcParams['image.cmap'] = 'gray'  # use grayscale output rather than a (potentially misleading) color heatmap
```

```python
def checkModelExist():
    # check if model exists
    if os.path.isfile(model_root + 'faceQuality.caffemodel'):
        print 'faceQuality found.'
    else:
        print 'faceQuality not found'

    if os.path.isfile(model_root + 'ImageQuality.caffemodel'):
        print 'ImageQuality found.'
    else:
        print 'ImageQuality not found'
```

```python
def loadAndResizeImage(image_path):
    img = cv2.imread(image_path, 0)
    longest_side = 480
    height, width = img.shape[:2]
    scale = 1.0 * longest_side / max(height, width)
    resized_image = cv2.resize(img, (0,0), fx = scale, fy = scale)
    return (resized_image, scale)
```

```python
def faceDetection(resized_image, scale):
    img = cv2.imread(img_path, 0)
    face_cascade = cv2.CascadeClassifier('./faceModel/face.xml')
    faces = face_cascade.detectMultiScale(resized_image, scaleFactor=1.1, minNeighbors=2,
                                             minSize=(30, 30), flags=0)
    for  (x, y, w, h) in faces:
        cv2.rectangle(resized_image, (x, y), (x + w, y + h), (0 , 255, 0), 2)
    
    plt.imshow(resized_image)
    return faces
```

```python
# image quality Set up model
def calculateImageQuality(image):
    caffe.set_mode_cpu()
    image_quality_model_def = model_root + 'ImageQuality.prototxt'
    image_quality_model_weights = model_root + 'ImageQuality.caffemodel'
    image_quality_net = caffe.Net(image_quality_model_def,      # defines the structure of the model
                image_quality_model_weights,  # contains the trained weights
                caffe.TEST)     # use test mode (e.g., don't perform dropout)
    mu = np.array([120.2599734, 129.036626861, 132.387003387])# BGR
    # create transformer for the input called 'data'
    transformer = caffe.io.Transformer({'data': image_quality_net.blobs['data'].data.shape})

    transformer.set_transpose('data', (2,0,1))  # move image channels to outermost dimension
    transformer.set_mean('data', mu)            # subtract the dataset-mean value in each channel
    transformer.set_raw_scale('data', 255)      # rescale from [0, 1] to [0, 255]
    transformer.set_channel_swap('data', (2,1,0))  # swap channels from RGB to BGR
    image = caffe.io.resize_image(image, (224,224), interp_order=3 )
    transformed_image = transformer.preprocess('data', image)
    image_quality_net.blobs['data'].data[...] = transformed_image
    output = image_quality_net.forward()
    print output
    image_quality = output['loss3/classifier_1'][0][0]
    return image_quality
```

```python
def calculateFaceQuality(image):
    #set up face quality
    # image quality Set up model
    caffe.set_mode_cpu()
    face_quality_model_def = model_root + 'faceQuality.prototxt'
    face_quality_model_weights = model_root + 'faceQuality.caffemodel'
    face_quality_net = caffe.Net(face_quality_model_def,      # defines the structure of the model
                    face_quality_model_weights,  # contains the trained weights
                    caffe.TEST)     # use test mode (e.g., don't perform dropout)
    mu = np.array([93.723474598, 107.165522511, 136.168507509])# BGR
    # create transformer for the input called 'data'
    transformer = caffe.io.Transformer({'data': face_quality_net.blobs['data'].data.shape})
    
    transformer.set_transpose('data', (2,0,1))  # move image channels to outermost dimension
    transformer.set_mean('data', mu)            # subtract the dataset-mean value in each channel
    transformer.set_raw_scale('data', 255)      # rescale from [0, 1] to [0, 255]
    transformer.set_channel_swap('data', (2,1,0))  # swap channels from RGB to BGR
    #image = caffe.io.load_image(img_root + 'groupSelfie.jpg')
    image = caffe.io.resize_image(image, (112,112), interp_order=3 )
    transformed_image = transformer.preprocess('data', image)
    face_quality_net.blobs['data'].data[...] = transformed_image
    output = face_quality_net.forward()
    face_quality = output['pool10'][0][0][0][0]
    return face_quality
```

```python
def calculateOneFrame(image_path, baseFaces):
    resized_image, scale = loadAndResizeImage(image_path)
    faces = faceDetection(resized_image, scale)
    
    print "baseFaces:"
    print baseFaces
    
    if (len(baseFaces) > 0):
        mappingRects = mypkg.CCBestFrameAlgorithm.mapFaceRectToBaseline(faces, baseFaces)
    
    currentFrameFaceQuality = [0] * len(baseFaces)
    
    image = caffe.io.load_image(image_path)
    cropped_faces = []
    for (x,y,w,h) in mappingRects:
        x = int(1.0 * x / scale)
        y = int(1.0 * y / scale)
        w = int(1.0 * w / scale)
        h = int(1.0 * h / scale)
        cropped_faces.append(image[y:y+h, x:x+w])
    
    if (len(cropped_faces) > 0):
        plt.imshow(cropped_faces[0])
   
    image_quality = calculateImageQuality(image)
    
    print len(cropped_faces)

    for i in range(len(cropped_faces)):
        face_quality = calculateFaceQuality(cropped_faces[i])
        print face_quality
        currentFrameFaceQuality[i] = face_quality  
    return (image_quality, currentFrameFaceQuality)
    
```

```python
def loadFramePaths():
    frame_root = './imgs/selfie_test'
    # frames starts in format "frame0001.jpg" fps = 10
    frame_names = os.listdir(frame_root)
    last_number = int(max(frame_names).lstrip('frame').strip(".jpg"))
    step = last_number / 10
    
    paths = []
    
    for i in range(10):
        cur_num = 1 + i * step
        transformed_num = str(cur_num)
        if (cur_num / 10 < 1):
            transformed_num = "000" + transformed_num
        elif (cur_num / 10 < 10):
            transformed_num = "00" + transformed_num
        elif (cur_num / 10 < 100):
            transformed_num = "0" + transformed_num
        paths.append(frame_root + "/"+ "frame" + transformed_num + ".jpg")
    return paths
        
```

```python
def main():
    setDisplay()
    checkModelExist()

    video_frame_paths = loadFramePaths()
    
    #tmp base faces
    resized_image, scale = loadAndResizeImage(video_frame_paths[0])
    plt.imshow(resized_image)
    baseFaces = faceDetection(resized_image, scale)
    
    print baseFaces
    
    _imageQualities = [-100] * len(video_frame_paths)
    _faceQualities = [[0] * len(baseFaces)] * len(video_frame_paths)
    
    bestFrameImageMode = "CCBestFrameImageModeNone"
    if (len(baseFaces) == 0):
        bestFrameImageMode = "CCBestFrameImageModeNone"
    elif (len(baseFaces) == 1):
        bestFrameImageMode = "CCBestFrameImageModeSingleFace"
    elif (len(baseFaces) > 1):
        bestFrameImageMode = "CCBestFrameImageModeMultiFaces"
    
    print bestFrameImageMode
    
    for i in range(len(video_frame_paths)):
        image_quality, currentFrameFaceQuality = calculateOneFrame(video_frame_paths[i], baseFaces)
        _imageQualities[i] = image_quality
        _faceQualities[i] = currentFrameFaceQuality
    
    scoreArray = mypkg.CCBestFrameAlgorithm.calculateBestFrameScoreArray(bestFrameImageMode, _imageQualities, _faceQualities, baseFaces)
    print scoreArray
    
    max_score = max(scoreArray)
    max_index = scoreArray.index(max_score)
    
    print "max_index: " + str(max_index)
    
    
```

```python
main()
```
